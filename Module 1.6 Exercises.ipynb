```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

df=pd.read_csv("datasets/tamiami.csv")


df=df.drop(0)


```

```python
df.corr()
```

```python
df.columns=['location','sales','employees','restaurants','foodcarts','price'] #sales, employees , restaurants, foodcarts seem to be correlated to price.
```

### Exercise 2:

Using the dataframe from the previous exercise, choose features (columns) to create a linear regression formula to predict sales. Try it with and without the y-intercept. How does it make a difference? Does adding or removing features in your model formula make a difference in the output?

```python
result=smf.ols('price ~ location + sales + employees + restaurants + foodcarts',data=df).fit()
```

```python
result.summary()
```

```python
result=smf.ols('price ~ location + sales + employees + restaurants + foodcarts-1',data=df).fit()
```

```python
result.summary()
```

```python
result_locationEmployeesRestCarts=smf.ols('price ~ location  + employees + restaurants + foodcarts-1',data=df).fit()
```
