### Exercise 4:                          

Load in the "travel_times.csv" file as a dataframe. Drop the "Comments" column. Then remove rows from the dataframe that have missing values and assign the resulting dataframe as a new variable.

```python
travel_data= pd.read_csv('datasets/travel-times.csv')
travel_data_noComments=travel_data.drop('Comments',axis=1)
travel_data_noComments=travel_data_noComments.dropna(axis=0)
travel_data_noComments.head(20)
```

### Exercise 5:

Using the dataframe from the exercise above (w/ no missing values), create bins that will categorize the AvgSpeed column as "slow" or "fast", and make a new column called "Speed" to hold those new values. Values less than 75 are "slow" and everything above is "fast".

```python
group_names=['slow','fast']
bins=[0,75,100]
travel_data_noComments["Speed"]=pd.cut(travel_data_noComments["AvgSpeed"],bins,labels=group_names)
travel_data_noComments.head()
```

### Exercise 7:

Using the dataframe from the previous exercise, pick a method (Standard Deviation or Interquartile Range) and remove the outliers from the "FuelEconomy" column.

```python
travel_data_noComments['FuelEconomy'].dtypes
```

```python
travel_data_clean=np.where(travel_data_noComments['FuelEconomy']=='-','0',travel_data_noComments['FuelEconomy'])
```

```python
travel_data_clean
```

```python
travel_data_noComments['FuelEconomy']=travel_data_clean
```

```python
travel_data_noComments.head()
```

```python
travel_data_noComments['FuelEconomy']=travel_data_noComments['FuelEconomy'].apply(lambda x:float(x))
```

```python
travel_data_noComments.head()
```

```python
meanval=travel_data_noComments['FuelEconomy'].mean()
```

```python
stdval=travel_data_noComments['FuelEconomy'].std()
```

```python
toprange=meanval+stdval*1.96
```

```python
botrange=meanval-stdval*1.96
```

```python
newdf=travel_data_noComments.copy()
```

```python
newdf=newdf.drop(newdf[newdf['FuelEconomy']>toprange].index)
```

```python
newdf=newdf.drop(newdf[newdf['FuelEconomy']<botrange].index)
```
